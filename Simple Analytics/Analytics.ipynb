```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
df=pd.read_csv(r'C:\Users\Prabashwara Randeni\Desktop\Data Analytics\UK Data Science Job & Salary Analytics\1_CSV Load\Cleaned_Data.csv')
```

```python
#inspec data
df.info()
```

```python
# Group by city and count job postings
city_job_count = df.groupby('City')['Job Title'].count().reset_index()

# Rename columns for clarity
city_job_count.columns = ['City', 'Job Count']

# Sort by Job Count to see the most active cities
city_job_count = city_job_count.sort_values(by='Job Count', ascending=False)

#----------------------------------------------------------------------------------------

# Group by country and count job postings
country_job_count = df.groupby('Country')['Job Title'].count().reset_index()

# Rename columns for clarity
country_job_count.columns = ['Country', 'Job Count']

# Sort by Job Count to see the most active countries
country_job_count = country_job_count.sort_values(by='Job Count', ascending=False)


#--------------------------------------------------------------------------------------------

# Group by city and calculate the average salary
city_salary_trend = df.groupby('City')['Salary'].mean().reset_index()

# Rename columns for clarity
city_salary_trend.columns = ['City', 'Average Salary']

# Sort by Average Salary to see cities with the highest salaries
city_salary_trend = city_salary_trend.sort_values(by='Average Salary', ascending=False)

```

```python
# Plot the top 10 cities with the most job postings
top_cities = city_job_count.head(5)

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Job Count', y='City', data=top_cities, palette='viridis')

# Add title and labels
plt.title('Top 5 Cities with Data Science Job Postings in the UK')
plt.xlabel('Number of Job Postings')
plt.ylabel('City')

# Show the plot
plt.show()
```

```python
# Plot the top countries with the most job postings
plt.figure(figsize=(10, 6))
sns.barplot(x='Job Count', y='Country', data=country_job_count, palette='coolwarm')

# Add title and labels
plt.title('Data Science Job Postings by Country')
plt.xlabel('Number of Job Postings')
plt.ylabel('Country')

# Show the plot
plt.show()
```

```python
# Plot the top 10 cities with the highest average salaries
top_salary_cities = city_salary_trend.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x='Average Salary', y='City', data=top_salary_cities, palette='magma')

# Add title and labels
plt.title('Top 5 Cities with Highest Average Salaries for Data Science Jobs')
plt.xlabel('Average Salary')
plt.ylabel('City')

# Show the plot
plt.show()
```

```python
# Group by country and calculate the average salary
country_salary_trend = df.groupby('Country')['Salary'].mean().reset_index()

# Rename columns for clarity
country_salary_trend.columns = ['Country', 'Average Salary']

# Sort by Average Salary to see countries with the highest salaries
country_salary_trend = country_salary_trend.sort_values(by='Average Salary', ascending=False)
```

```python
# Plot the top 10 countries with the highest average salaries
top_salary_countries = country_salary_trend.head()

plt.figure(figsize=(10, 6))
sns.barplot(x='Average Salary', y='Country', data=top_salary_countries, palette='cool')

# Add title and labels
plt.title('Top Countries with Highest Average Salaries for Data Science Jobs')
plt.xlabel('Average Salary')
plt.ylabel('Country')

# Show the plot
plt.show()
```
